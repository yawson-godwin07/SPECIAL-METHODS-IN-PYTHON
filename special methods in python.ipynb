# 1. Constructors in Python (`__init__()`)

## Definition
The `__init__()` method is a special method (constructor) that is automatically called whenever an object is created from a class.

It is mainly used to initialize the attributes of an object, ensuring that every object starts with the required values.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Matthew", 19)
print(s1.name)
print(s1.age)
```

# 2. String Representation (`__str__()`)

## Definition
The `__str__()` method provides a human-readable representation of an object.

Whenever the `print()` function is used on an object, Python automatically calls the `__str__()` method. Without this method, Python displays the object's memory location, which is not user-friendly.

```python
class StudentStr:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student: {self.name}"

s2 = StudentStr("Matthew")
print(s2)
```

# 3. Formal Representation (`__repr__()`)

## Definition
The `__repr__()` method provides an official or developer-oriented representation of an object.

It is mainly used for debugging and logging. If the `__str__()` method is not defined, Python automatically uses the `__repr__()` method as a backup.

```python
class StudentRepr:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f"Student('{self.name}')"

s3 = StudentRepr("Daniella")
print(repr(s3))
```

# 4. Length Method (`__len__()`)

## Definition
The `__len__()` special method allows a custom object to define its own length. It enables Python's built-in `len()` function to work with user-defined classes.

```python
class Playlist:
    def __init__(self, songs):
        self.songs = songs

    def __len__(self):
        return len(self.songs)

p = Playlist(["Song1", "Song2", "Song3"])
print(len(p))
```

# 5. Equality Comparison (`__eq__()`)

## Definition
The `__eq__()` method defines how two objects should be compared for equality using the `==` operator. Instead of comparing memory addresses, it compares the actual contents of the objects.

```python
class Box:
    def __init__(self, size):
        self.size = size

    def __eq__(self, other):
        if isinstance(other, Box):
            return self.size == other.size
        return False

b1 = Box(10)
b2 = Box(10)
print(b1 == b2)
```

# 6. Addition (`__add__()`)

## Definition
The `__add__()` method overloads the addition (`+`) operator for custom objects. It allows objects to be added together in a meaningful way.

```python
class VectorAdd:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return VectorAdd(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = VectorAdd(2, 3)
v2 = VectorAdd(4, 5)
print(v1 + v2)
```

# 7. Subtraction (`__sub__()`)

## Definition
The `__sub__()` method overloads the subtraction (`-`) operator for custom objects, enabling algebraic manipulation between classes.

```python
class VectorSub:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __sub__(self, other):
        return VectorSub(self.x - other.x, self.y - other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v3 = VectorSub(10, 8)
v4 = VectorSub(3, 2)
print(v3 - v4)
```

# 8. Multiplication (`__mul__()`)

## Definition
The `__mul__()` method defines how the multiplication (`*`) operator works with custom objects. It makes user-defined classes behave like built-in numeric types.

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __mul__(self, other):
        return Number(self.value * other.value)

    def __str__(self):
        return str(self.value)

n1 = Number(5)
n2 = Number(4)
print(n1 * n2)
```

# 9. Index Access (`__getitem__()`)

## Definition
The `__getitem__()` method allows an object to use bracket indexing notation like lists or dictionaries (e.g., `object[index]`).

```python
class RecordBook:
    def __init__(self, records):
        self.records = records

    def __getitem__(self, index):
        return self.records[index]

book = RecordBook(["Record A", "Record B", "Record C"])
print(book[1])
```

# 10. Index Assignment (`__setitem__()`)

## Definition
The `__setitem__()` method allows you to update or assign values using index positioning syntax (e.g., `object[index] = value`).

```python
class Grades:
    def __init__(self):
        self.data = [85, 90, 78]

    def __setitem__(self, index, value):
        self.data[index] = value

g = Grades()
g[1] = 100
print(g.data)
```

# 11. Membership Testing (`__contains__()`)

## Definition
The `__contains__()` method defines how the `in` operator behaves for custom objects. It checks whether a particular item exists inside the object.

```python
class Course:
    def __init__(self):
        self.students = ["Ella", "Sandra", "Valerie"]

    def __contains__(self, student):
        return student in self.students

c = Course()
print("Joe" in c)
print("Ella" in c)
```

# 12. Callable Objects (`__call__()`)

## Definition
The `__call__()` method allows an instance of a class to be called completely like a normal function using parentheses (e.g., `object()`).

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, value):
        return value * self.factor

triple = Multiplier(3)
print(triple(10))
```

# 13. Iterators (`__iter__()` and `__next__()`)

## Definition
The `__iter__()` method returns the iterator object itself, while the `__next__()` method returns the next item in the sequence. Together, they allow an object to be used inside a standard `for` loop.

```python
class Counter:
    def __init__(self, low, high):
        self.current = low
        self.high = high

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.high:
            raise StopIteration
        else:
            self.current += 1
            return self.current - 1

for num in Counter(1, 3):
    print(num)
```

# 14. Boolean Evaluation (`__bool__()`)

## Definition
The `__bool__()` method determines whether an object evaluates to `True` or `False`. It is automatically used in Boolean expressions and conditional `if` statements.

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def __bool__(self):
        return self.balance > 0

account = BankAccount(500)
if account:
    print("Account is active")
```

# 15. Destructor (`__del__()`)

## Definition
The `__del__()` method is called a destructor. It is invoked automatically when an object is about to be destroyed or garbage collected, allowing for resource cleanup.

```python
class Resource:
    def __init__(self):
        print("Resource initialized.")

    def __del__(self):
        print("Resource clean up complete. Instance destroyed.")

r = Resource()
del r
```
